# Affichage des résultats de classification :

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## Tableau à double entrées Position_GJ / Parti politique : 

On veut classifier la proportion de chaque parti à être gilet jaune ou non.

Fonction pour afficher le résultat : 

```python
def tableau_double_entrée(labels_position_gj,labels_partis):
    #Initialisation de la Figure : 
    fig=plt.figure(figsize = (10,7))  
    d={'Partis' : labels_partis.party, 'Position_Gilets_Jaunes' : labels_position_gj.position_gj}
    data=pd.DataFrame(data=d)
    data_crosstab = pd.crosstab(data['Partis'], data['Position_Gilets_Jaunes'],normalize=True) 
    #Afficher en pourcentages
    sns.heatmap(data_crosstab, annot=True, fmt='.2f', cmap="Greens")
    plt.title('Classification du parti et de la position GJ')
    plt.ylabel('Partis')
    plt.xlabel('Position Gilets Jaunes')
    plt.show()
    return True
```
